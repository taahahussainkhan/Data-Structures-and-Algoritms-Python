## Understanding figures

```python
%matplotlib notebook
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(3)
plt.plot(x,x)
plt.plot(x,2*x)
plt.plot(x,3*x)
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,5,5)
y=2*x
plt.plot(x,y)
#plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
plt.show()
```
